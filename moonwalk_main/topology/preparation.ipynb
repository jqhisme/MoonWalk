```python
json_dir = 'C:\\Users\\jiang\\Desktop\\NYU ITP\\Thesis\\Odyssey\\topology_of_storytelling\\embeddings\\json'
for index,f in enumerate(emb_files):
    data = np.load(os.path.join(emb_dir, f))
    features_3d = data['features']

    reducer = UMAPWrapper(n_neighbors=15, min_dist=0.1, n_components=3)

    features_3d = reducer.wrapper_fit_transform(features_3d)
    print(features_3d)

    # write the features in a json file, the metadata contains the name of the video and the variance of the features
    # and the data is the features_3d
    metadata = {
        "video_name": f.split('_')[0],
        "variance": round(float(np.var(features_3d, axis=0)[0]), 4)
    }

    content = {
        "metadata": metadata,
        "data": features_3d.tolist()}
    
    with open(os.path.join(json_dir, f"{f.split('.')[0]}.json"), 'w') as outfile:
        json.dump(content, outfile)
```

```python
import matplotlib.pyplot as plt
import re

# Function to clean and format video names
def clean_video_name(filename):
    """Remove special characters and split long names into two lines"""
    # Remove file extension if present
    name = filename.replace('_reduced.npz', '').replace('.npz', '')
    
    # Remove special characters but keep spaces
    name = re.sub(r'[│｜：:|−\-—]', ' ', name)
    name = re.sub(r'[()]', '', name)
    
    # Clean up extra spaces
    name = re.sub(r'\s+', ' ', name).strip()
    
    # Split into two lines if too long (>30 chars)
    if len(name) > 30:
        words = name.split()
        mid = len(words) // 2
        name = ' '.join(words[:mid]) + '\n' + ' '.join(words[mid:])
    
    return name

fig = plt.figure(figsize=(15, 10))
for index in range(len(vid_files)):
    data = np.load(os.path.join(output_dir, f"{vid_files[index]}_reduced.npz"))
    features_3d = data['features']
    ax = fig.add_subplot(2, 3, index+1, projection='3d')
    ax.scatter(features_3d[:, 0], features_3d[:, 1], features_3d[:, 2], s=50)
    clean_title = clean_video_name(vid_files[index])
    ax.set_title(clean_title, fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# for all the videos in /videos directory, extrac 6 frames from each of them, and concat them in 3 col 2 row, and save the rsult in 
# previewImg directory with the name of the video as the name of the image
vid_dir = 'C:\\Users\\jiang\\Desktop\\NYU ITP\\Thesis\\Odyssey\\topology_of_storytelling\\videos'
preview_img_dir = 'C:\\Users\\jiang\\Desktop\\NYU ITP\\Thesis\\Odyssey\\topology_of_storytelling\\previewImg'

import cv2
import numpy as np

for index,vid in enumerate(vid_files):
    vid_path = os.path.join(vid_dir, vid)
    cap = cv2.VideoCapture(vid_path)
    
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    offset = 10  # to avoid the very first frames which are often black
    frame_indices = np.linspace(offset, frame_count-offset, 6, dtype=int)

    frames = []
    for idx in frame_indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ret, frame = cap.read()
        if ret:
            frames.append(frame)
    
    cap.release()

    # Concatenate frames into a 3x2 grid
    top_row = np.hstack(frames[:3])
    bottom_row = np.hstack(frames[3:])
    preview_img = np.vstack((top_row, bottom_row))

    # Save the preview image
    preview_img_path = os.path.join(preview_img_dir, f"{vid_names[index]}_preview.jpg")
    cv2.imwrite(preview_img_path, preview_img)
```
